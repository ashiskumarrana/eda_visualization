```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="quality", y="pH")
plt.title("pH Level Distribution by Quality Score")
plt.xlabel("Wine Quality Score")
plt.ylabel("pH Level")
plt.show()
```

```python


# Calculate Q1 and Q3
Q1 = df["pH"].quantile(0.25)
Q3 = df["pH"].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Extract outliers
outliers = df[(df["pH"] < lower_bound) | (df["pH"] > upper_bound)]
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["pH"])
sns.stripplot(y=outliers["pH"], color='red', alpha=0.6)  # highlights outliers
plt.title("pH Outliers Highlighted")
plt.ylabel("pH Level")
plt.show()
```
